```python
# Fonction qui crée la liste de couleurs en fonction de la condition
def donner_couleur(x,y):
    couleurs=[]
    for i in range(len(x)):
        if y[i]<x[i]**2 : 
            couleurs.append("r")
        else : 
            couleurs.append("b")
    return couleurs
    
x = [random() for _ in range(5000)]
y = [random() for _ in range(5000)]
plt.scatter(x,y,s=1,c=donner_couleur(x,y))

plt.show()
```

On peut se servir de ce type de méthode pour donner une approximation de $\pi$.

En traçant la courbe d'équation $y=1-\sqrt{1-x^2}$ pour $0\leq x\leq 1$, on obtient un quart de cercle de centre le point de coordonnées  $(0;1)$ et de rayon 1.

La zone située au-dessus de cette courbe en restant à l'intérieur du carré définissant l'unité d'aire a donc pour aire $\dfrac{\pi}{4}$.

En plaçant aléatoirement des points dans ce carré et en calculant la fréquence des points situés au-dessus de la courbe, on a donc une approximation de $\dfrac{\pi}{4}$.

Ainsi, en multipliant par 4, on obtient une approximation de $\pi$.

```python
N = 0 #compteur des points situés au-dessus de la courbe
points = 1000 #nombre de points à placer : modifiable
for i in range(points):
    x, y = random(), random()
    if  y >= 1-math.sqrt(1-x**2):
        plt.scatter(x, y, c="r", marker="s") # couleur "red" et marqueur "square"
        N += 1
    else:
        plt.scatter(x, y, c="b", marker="o") # couleur "blue" et marqueur cercle
plt.show
print("pi est environ égal à : ",N/points*4)
```
